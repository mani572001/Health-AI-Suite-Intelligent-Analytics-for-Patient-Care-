```python
# =========================================
# 06_association_rules.ipynb
# Step 1: Imports
# =========================================

import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# =========================================
# Step 2: Load Data
# =========================================

df = pd.read_csv("../data/processed/cleaned_data.csv")

print(df.shape)
```

```python
# =========================================
# Step 3: Prepare Transaction Data
# =========================================

transaction_cols = [
    "diabetes",
    "hypertension",
    "heart_disease",
    "metabolic_risk"
]

transactions = df[transaction_cols].copy()

transactions.head()
```

```python
# =========================================
# Step 4: Run Apriori
# =========================================

frequent_itemsets = apriori(
    transactions,
    min_support=0.05,   # 5% minimum support
    use_colnames=True
)

frequent_itemsets.sort_values("support", ascending=False).head()
```

```python
# =========================================
# Step 5: Generate Association Rules
# =========================================

rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1.0
)

rules.sort_values("lift", ascending=False).head(10)
```

```python
rules.to_csv("../data/processed/association_rules.csv", index=False)
print("Association rules saved.")
```
